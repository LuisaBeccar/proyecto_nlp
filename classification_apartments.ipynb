# Text Classification

mi dataset es bastante chico: 101 filas, asique separarlo en train y test seria achicar mas la muestra de training... pero si lo hiciera, seria algo asi:

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/LuisaBeccar/proyecto_nlp/refs/heads/main/comentarios_departamentos.csv")
```

```python
X = df2.comentario
y = df2.balcon
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
"""
%%capture
!pip install nltk
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
"""
```

```python
# Initialize the Count Vectorizer
spanish_stop_words = stopwords.words('spanish')
count_vectorizer2 = CountVectorizer(stop_words=spanish_stop_words)

# Fit the model and transform the data
X_train_v = count_vectorizer2.fit_transform(X_train)


clf = LogisticRegression()
clf.fit(X_train_v, y_train)
```

```python
X_test.shape
```

```python
X_test_v = count_vectorizer2.transform(X_test)
```

```python
y_pred = clf.predict(X_test_v)
```

y se podria calcular la accuracy del modelo

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```

mediocre, posiblemente por la redicida cantida de casos de entrenamiento

#TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words=spanish_stop_words)

# Transform the text data
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
X_train_tfidf
```

```python
X_train_tfidf.toarray()
```

```python
clf3 = LogisticRegression()
clf3.fit(X_train_tfidf, y_train)
```

```python


# Make predictions
y_pred3 = clf3.predict(X_test_tfidf)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred3)
print(f"Accuracy: {accuracy:.4f}")
```
